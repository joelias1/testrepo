```python
# Dependencies and Setup
import pandas as pd

# Create reference for the csv file
purchase_data = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(purchase_data)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Displaying total number of players:
TotalPlayers = len(purchase_data["SN"].unique())
totalplayer = pd.DataFrame({"Total Players" : [TotalPlayers]})
totalplayer.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate Purchasing Analysis(Total)
# Data Input
unique_items = purchase_data["Item ID"].nunique()
unique_items
average_price = round(purchase_data["Price"].mean(), 2)average_price
purchase_number = purchase_data["Item ID"].count()
purchase_number
total_revenue = purchase_data["Price"].sum()
total_revenue

# Creating a summary dataframe:
purchase_summary = pd.DataFrame({"Number of Unique Items" : [unique_items],
                                 "Average Price" : [average_price], 
                                 "Number of Purchases" : [purchase_number], 
                                 "Total Revenue" : [total_revenue]})

purchase_summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Analyzing Gender Demographics
df = purchase_data.drop_duplicates(subset='SN', keep="first")
total_count = df["Gender"].count()
total_male = df["Gender"].value_counts()['Male']
total_female = df["Gender"].value_counts()['Female']
gender_other = total_count - total_male - total_female

#Calculating the percentage count, Data input
male_count = (total_male / total_count) * 100
female_count = (total_female / total_count) * 100
gender_none = (gender_other / total_count) * 100
# Create DataFrame
gender_summary = pd.DataFrame({"": ['Male', 'Female', 'Other/Non-Disclosed'],
                              "Percentage of Players": [male_count, female_count, gender_none],
                              "Total Counts": [total_male, total_female, gender_other]})


# Formatting DataFrame
gender_summary["Percentage of Players"] = gender_summary["Percentage of Players"].map("{:.2f}%".format)

gender_summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# CALCULATING PURCHASING ANALYSIS (GENDER)
# Group by gender & drop duplicates entry
group_gender = purchase_data.groupby(["Gender"])
print(group_gender)

# Deleting duplicates
df = purchase_data.drop_duplicates(subset='SN', keep="first")
group_df = df.groupby(["Gender"])

#data input
purchase_count = group_gender["SN"].count()
purchase_price = group_gender["Price"].mean()
purchase_total = group_gender["Price"].sum()

#Find purchase per person
purchase_df = round(group_gender["Price"].sum() / group_df["SN"].count(),2)
                              
# Create DataFrame analysis_summary 
analysis_summary = pd.DataFrame({"Purchase Count": purchase_count,
                                 "Average Purchase Price": purchase_price,
                                 "Total Purchase Value": purchase_total,
                                 "Avg Total Purchase per Person": purchase_df})
#Format DataFrame
analysis_summary["Average Purchase Price"] = analysis_summary["Average Purchase Price"].map("${:.2f}".format)
analysis_summary["Total Purchase Value"] = analysis_summary["Total Purchase Value"].map("${:.2f}".format)
analysis_summary["Avg Total Purchase per person"] = analysis_summary["Avg Total Purchase per Person"].map("${:.2f}".format)
analysis_summary = analysis_summary[["Purchase Count", 
                                     "Average Purchase Price", 
                                     "Total Purchase Value", 
                                     "Avg Total Purchase per Person"]]
analysis_summary            



                              
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# CALCULATING AGE DEMOGRAPHICS
# Binning
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

#Create names for bins
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

#Adding bins to DataFrame
bin_names = purchase_data.copy()
bin_names["Age Groups"] = pd.cut(bin_names["Age"], bins, labels=group_names)
bin_group = bin_names.groupby(["Age Groups"])
bin_group

# Create Data input
count_df = bin_group["SN"].count()
count_bins = purchase_data["SN"].count()
percentage_group = (count_df / count_bins) * 100
percentage_group

# Creating the DataFrame
demographic_summary = pd.DataFrame({"Total Count": count_df,
                                    "Percentage of Players": percentage_group})  

# Formating DataFrame percentage
demographic_summary["Percentage of Players"] = demographic_summary["Percentage of Players"].map("{:.2f}%".format)
                    
demographic_summary





```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# CALCULATING AGE DEMOGRAPHICS

bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

#Create names for bins
age_data = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

#Adding bins to DataFrame
age_names = purchase_data.copy()
age_names["Age Groups"] = pd.cut(age_names["Age"], bins, labels=age_data)
age_group = age_names.groupby(["Age Groups"])
age_group

# Create Data input
count_df = age_group["SN"].count()
count_bins = purchase_data["SN"].count()

avg_price = round(age_group["Price"].mean(), 2)
avg_price

age_value = age_group["Price"].sum()
age_value


purchase_df = round(age_group["Price"].sum() / age_group["SN"].count(),2)
purchase_df

# Creating the DataFrame
age_summary = pd.DataFrame({"Purchase Count": count_df,
                            "Average Purchase Price": avg_price,
                            "Total Purchase Value": age_value,
                            "Avg Purchase per Person": purchase_df})  

# Formating DataFrame percentage
age_summary["Average Purchase Value"] = age_summary["Average Purchase Price"].map("${:.2f}".format)
age_summary["Total Purchase Value"] = age_summary["Total Purchase Value"].map("${:.2f}".format)                    
age_summary["Avg Purchase per Person"] = age_summary["Avg Purchase per Person"].map("${:.2f}".format)                    

age_summary



```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Data Input for Calculation of Top Spenders
group_spenders = purchase_data.groupby(["SN"])
spenders_count = group_spenders["Item ID"].count()
spenders_total = group_spenders["Price"].sum()
spenders_avg = (spenders_total / spenders_count)

# New DataFrame
spenders_summary = pd.DataFrame({"Purchase Count": spenders_count,
                                "Average Purchase Price": spenders_avg,
                                "Total Purchase Value": spenders_total})

# Formatting DataFrame Result

spenders_summary = spenders_summary.sort_values("Total Purchase Value", ascending=False)
spenders_summary["Average Purchase Price"] = spenders_summary["Average Purchase Price"].map("${:.2f}".format)
spenders_summary["Total Purchase Value"] = spenders_summary["Total Purchase Value"].map("${:.2f}".format)

spenders_summary.head()            

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Data Input for Calculation of Most Popular Items
group_popular = purchase_data.groupby(["Item ID", "Item Name"])
popular_count = group_popular["SN"].count()
popular_total = group_popular["Price"].sum()
popular_price = (popular_total / popular_count)
popular_value = (popular_count / popular_price)

# New DataFrame
popular_summary = pd.DataFrame({"Purchase Count": popular_count,
                                "Item Price": popular_price,
                                "Total Purchase Value": popular_value })

# Formatting DataFrame Result

popular_summary = popular_summary.sort_values("Purchase Count", ascending=False)
popular_summary["Item Price"] = popular_summary["Item Price"].map("${:.2f}".format)
popular_summary["Total Purchase Value"] = popular_summary["Total Purchase Value"].map("${:.2f}".format)
popular_summary = popular_summary[["Purchase Count", "Item Price", "Total Purchase Value"]]
popular_summary.head()            

```
