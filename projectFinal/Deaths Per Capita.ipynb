```python
import pandas as pd
import requests
import json
from config import gkey
```

```python
# Create dataframe from cleaned data
clean_data = pd.read_csv("Resources/DrugsbyCityWLatLng.csv")
clean_data.head()
```

```python
# Group data by Death City by count
city_counts = clean_data.loc[:,['City','ID','Latitude','Longitude']]
city_counts = city_counts.rename(columns={"ID":"Deaths"})
city_counts.head()
```

```python
# Read in town population data
pop_towns2018 = pd.read_csv("Resources/pop_towns2018.csv",header=None)
pop_towns2018 = pop_towns2018.rename(columns={0:'City',1:'Pop'})
pop_towns2018['City'] = pop_towns2018['City'].str.upper()
pop_towns2018['Pop'] = pop_towns2018['Pop'].str.replace(',','')
pop_towns2018.head()
```

```python
city_merged = city_counts.merge(pop_towns2018,on='City')
city_merged['Pop'] = city_merged['Pop'].astype('int64')
city_merged.head()
```

```python
city_merged.count()
```

```python
city_merged['PerCapita'] = ''
for index,row in city_merged.iterrows():
    temp = float(row['Deaths']) / float(row['Pop']) * 100000
    # per 100000 people
    city_merged.loc[index,'PerCapita'] = temp
    
city_merged.head()
```

```python
import gmaps
gmaps.configure(api_key=gkey)
```

```python
ct_coord = (41.5032, -72.6877)
capita_map = gmaps.figure(center=ct_coord,zoom_level=8.4)
heatmap_layer = gmaps.heatmap_layer(city_merged[['Latitude','Longitude']],
                                   weights=city_merged['PerCapita'],
                                   point_radius=40.0)
capita_map.add_layer(heatmap_layer)
capita_map
```

```python
bins = [0,10000,50000,100000,150000]
bin_labels = ['0-10000','10000-50000','50000-100000','100000-150000']
city_merged['PopBin'] = pd.cut(city_merged['Pop'],bins,labels=bin_labels)
city_merged
```

```python
city_binned = city_merged.groupby('PopBin').mean()
city_binned['N'] = city_merged.groupby('PopBin').count()['City']
city_binned['PerCapita'] = city_binned['Deaths'] / city_binned['Pop'] * 100000
city_binned
```

```python
per_capita_bar = city_binned['PerCapita'].plot(kind='bar',figsize=(9,6))
per_capita_bar.set_title('Deaths per Capita by City Size')
per_capita_bar.set_xlabel('City Population')
per_capita_bar.set_ylabel('Deaths per 100K People')
per_capita_bar.set_xticklabels(city_binned.index,rotation=0)
for i, label in enumerate(list(city_binned['N'])):
    temp = 'N='+str(label)
    per_capita_bar.annotate(temp,(i-0.1,50))
# Export Figure
per_capita_bar.figure.tight_layout()
per_capita_bar.figure.savefig('Charts/DeathsPerCapita.png')
per_capita_bar
```
