```python
import pandas as pd
import requests
import json
import ipywidgets as widgets
from colormap import rgb2hex
from config import gkey
import gmaps
#pd.set_option('display.max_rows', None)
```

```python
clean_data = pd.read_csv("Resources/drug_death_data_clean.csv")
#clean_data
```

```python
dayOfWeekDF = pd.read_csv("Resources/dayOfWeekDeaths.csv")
clean_data["Day of Week"] = dayOfWeekDF["dayOfWeek"]
clean_data.head()
```

```python
#groupedbyCityDay
def dayColor(day):
    if day == "Friday":
        color = '#E9FF33'
    if day == "Saturday":
        color = '#1F9E0B'
    if day == "Sunday":
        color = '#F6BD0A'
    if day == "Monday":
        color = "#F61F0A"
    
    if day == "Tuesday":
        color = "#000002"
    
    if day == "Wed":
        color = "#0A0AF6"
    
    if day == "Thurs":
        color = '#FA0BFA'
    return color
       

color = [dayColor(row["Day of Week"]) for index,row in clean_data.iterrows() ]
clean_data["Color"] = pd.DataFrame(color)


cityDayGrouped = clean_data.groupby(["Day of Week","DeathCity","Color"])
cityCountDay = pd.DataFrame(cityDayGrouped.count()["ID"])
cityCountDay = cityCountDay.reset_index()

cityCountDay = cityCountDay.rename(columns={"DeathCity":"City","ID":"Deaths"})
cityCountDay = cityCountDay.loc[cityCountDay["City"] != "06340",:]
(cityCountDay)
```

```python
# Create new columns for city coordinates
#using Josh's version
cityCountDay["Latitude"] = ""
cityCountDay["Longitude"] = ""


# Call google places api for each city and retrieve coordinates
base_url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json?"
params = {"key":gkey,
         "input":"",
         "inputtype":"textquery",
         "fields":"geometry/location"}

for index, row in cityCountDay.iterrows():
    # Change input for each row/city.  Add "CT" since we're working in Connecticut
    params["input"] = row["City"] + ",CT"
    # Call google places API
    response = requests.get(base_url,params)
    print(f"Processing {row['City']}...")
    # If good response, store lat and lng in df
    if response.status_code == 200:
        try:
            cityCountDay.loc[index,"Latitude"] = response.json()['candidates'][0]["geometry"]["location"]["lat"]
            cityCountDay.loc[index,"Longitude"] = response.json()['candidates'][0]["geometry"]["location"]["lng"]
        except IndexError:
          print(f"Information not found for {row['City']}! Continuing...")
    else:
        print(f"Information not found for {row['City']}! Continuing...")
              
print("Latitude/Longitude processing complete!")
```

```python
#modified from https://github.com/pbugnion/gmaps/issues/210

import time #attempt to slow showing of different marker layers(failed do to nature of gmaps functionality)
from matplotlib.cm import tab10 
from matplotlib.colors import to_hex

import ipywidgets as widgets

gmaps.configure(api_key=gkey)

    
days = ['Monday', "Sunday", 'Wed', 'Thurs', 'Friday', 'Saturday','Tuesday']




data = {day: getZippedLocations(day) for day in days }
#print(data)
colors = {day: to_hex(tab10(iday)) for iday, day in enumerate(days)} # generate a colour for each year

# Some CSS hackery to draw a circle of that color next to the year
legend_template = '<span style="color: {color}; font-size: 2em; vertical-align: bottom;">&#9679;</span> {day}'
ct_coord = (41.5032, -72.6877)
figure = gmaps.figure(center=ct_coord, zoom_level=8.4)

#def draw_point_layers():
    
legend_entries = []
pointsLayers = []
for iday, day in enumerate(days):
    locations = data[day]
    color = colors[day]
    #current = cityCountDay[cityCountDay["Day of Week"]=="day"]
    #sizeMod = len(current[current["Longitude"]==locations[1] ])
    points_layer = gmaps.symbol_layer(
        locations, 
        fill_color=color,
        scale=4,
        stroke_color=color, 
        fill_opacity=0.3,
        stroke_opacity=1
        
  
    )
    pointsLayers.append(points_layer)
    legend_entries.append(legend_template.format(color=color, day=day))
legend = widgets.HBox(
        [widgets.HTML(legend_entry) for legend_entry in legend_entries],
        layout=widgets.Layout(width='100%', justify_content='space-between')
    )

    

for i in range(7):
    figure.add_layer(pointsLayers[i])

widgets.VBox([figure, legend])
```
